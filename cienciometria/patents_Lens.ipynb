````python
df=pd.DataFrame()
for p in soup.find_all('div',{'class':"div-table-results-row no-selects ng-scope"}):
    d={}
    break
    #Extraer metadatos de cada patente. En este caso, para el título de la patente 
    nt = p.find('h3', {'style':'font-size: 1rem;'})
    nd = p.find('ul', {'class':'list-inline header-meta'})
    np = p.find('div', {'class':'listing-sidebar results-doc-id'})
    npl = np.find('div',{'class':'doc-type'})
    np.prettify
    #Use strip para remover espacios en blanco y rupturas de líneas al comienzo o al final de la cadena de texto
    try:
        d['title'] = nt.find('a').contents[0]
        d['publication_date'] = nd.find('li').contents[1]
        d['type'] = np.find('h3').contents[0]
        d['link'] = npl.find('a')
    except IndexError:
        d['title']=''
        d['publication_date']=''
        d['type'] = ''
    #Extraer fecha de publicación de la patente
    
    #Extraer Aplicante de la patente
    d['applicant'] = nd.find('li').contents[-2]
    print(d)
````

````python
df=pd.DataFrame()
fc=soup.find_all('div',{'class':"div-table-results-row no-selects ng-scope"})
for p in fc:
    d={}
    da=p.findAll('li',{'class':'ng-binding'})
    for j in da:
        l=j.text.split(': ')
        try:
            d[l[0]]=l[1]
        except IndexError:
            pass
        
    da=p.findAll('a',{'class':'ng-binding'})
    da=[ x for x in da if x.get('href')] # Filtre los tag a que contengan href
    d['id']=da[1].text.strip()
    d['link']=da[2].text.strip()
    df=df.append(d,ignore_index=True)    
````
